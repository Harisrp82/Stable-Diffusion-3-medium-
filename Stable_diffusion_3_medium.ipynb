```python
import torch
from diffusers import StableDiffusion3Pipeline
from transformers import T5EncoderModel, BitsAndBytesConfig

quantization_config =BitsAndBytesConfig(load_in_8bit=True)

model_id="stabilityai/stable-diffusion-3-medium-diffusers"
text_encoder=T5EncoderModel.from_pretrained(
    model_id,
    subfolder="text_encoder_3",
    quantization_config =quantization_config,
)
pipe=StableDiffusion3Pipeline.from_pretrained(
    model_id,
    text_encoder_3=text_encoder,
    device_map="balanced",
    torch_dtype=torch.float16
)

def create_image(prompt, negative_prompt="", num_inference_steps=28, height=1024, width=1024, guidance_scale=7.0):
    # Ensure height and width are divisible by 8
    height = (height // 8) * 8
    width = (width // 8) * 8

    try:
        # Convert prompt and negative_prompt to lists if they aren't already
        if isinstance(prompt, str):
            prompt = [prompt]
        if isinstance(negative_prompt, str):
            negative_prompt = [negative_prompt]

        image = pipe(
            prompt=prompt,
            negative_prompt=negative_prompt,
            num_inference_steps=num_inference_steps,
            height=height,
            width=width,
            guidance_scale=guidance_scale
        ).images[0]

        return image
    except Exception as e:
        print(f"Error creating image: {e}")
        return None

```

```python
image1=create_image("the man wearing jacket and walking through the mars","mars.png")
```

```python
import gradio as gr

# Define the Gradio interface with simplified components
with gr.Blocks() as interface:
    gr.Markdown("# Stable Diffusion-3-medium-Image Generator")

    with gr.Row():
        prompt = gr.Textbox(label="Prompt")
        negative_prompt = gr.Textbox(label="Negative Prompt (optional)")

    generate_button = gr.Button("Generate Image")
    output_image = gr.Image(type="pil", label="Generated Image")

    # Set the function to run on button click
    generate_button.click(
        create_image,
        inputs=[prompt, negative_prompt],
        outputs=output_image
    )

# Launch the interface
interface.launch(debug=True, share=True)
```
